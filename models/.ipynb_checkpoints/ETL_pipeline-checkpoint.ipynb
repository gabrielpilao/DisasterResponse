```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
#importing files
messages_df = pd.read_csv('messages.csv')
categories_df = pd.read_csv('categories.csv')
```

```python
messages_df.head()
```

```python
categories_df.head()
```

```python
#joining dataframes
joint_df = messages_df.merge(categories_df, on=['id'])
```

```python
joint_df.head()
```

```python
joint_df_categories = joint_df['categories'].str.split(';', expand = True)
```

```python
joint_df_categories.head()
```

```python
row = joint_df_categories.iloc[0, :].copy()
```

```python
col_names = row.apply(lambda x: x[:-2])
```

```python
joint_df_categories.columns = col_names
joint_df_categories.head()
```

```python
for column in joint_df_categories:
    joint_df_categories[column] = pd.to_numeric(joint_df_categories[column])
joint_df_categories.head()
```

```python
df_concat = pd.concat([joint_df, joint_df_categories], axis = 1)
```

```python
df_concat
```

```python
df = df_concat.iloc[:, 5:].apply(lambda x: x.str.split('-', expand=True)[1])
```

```python
df.head()
```

```python
df_final2 = df_concat.drop(df_concat.columns[4:], axis=1)
```

```python
df_final = pd.concat([df_final2, df], axis = 1) 
```

```python
df_final.head()
```

```python
df_final['message'].nunique()
```

```python
df_final.drop_duplicates(subset = 'message', inplace= True)
```

```python
df_final.shape
```

```python
engine = create_engine('sqlite:///etl_figure_eight.db')
df_final.to_sql('etl_figure_eight', engine, index=False)
```
